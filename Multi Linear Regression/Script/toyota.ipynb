# Multilinear Regression Model

```python
import statsmodels.formula.api as smf

ml1 = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data=tcar).fit()  # regression model
ml1.params
```

```python
ml1.summary()
```

```python
ml1_pred = ml1.predict(tcar)
ml1_error = ml1_pred-tcar.Price
print(np.sqrt(np.mean(ml1_error*ml1_error)),"RMSE value of ML1") 
```

```python
import statsmodels.api as sm
sm.graphics.influence_plot(ml1)
plt.show()
```

```python
tcar_new=tcar.drop(tcar.index[[80,221,601,960]],axis=0)  
```

```python
# preparing model
ml_new = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data=tcar_new).fit()
ml_new.summary()
```

```python
mlnew_pred = ml_new.predict(tcar_new)
mlnew_rmse = mlnew_pred-tcar_new.Price
print(np.sqrt(np.mean(mlnew_rmse*mlnew_rmse)),"RMSE value of MLnew") 
```

```python
# Observed values VS Fitted values
plt.scatter(tcar_new.Price,mlnew_pred,c="r");plt.xlabel("observed_values");plt.ylabel("fitted_values")
plt.show()
```

```python
# Residuals VS Fitted Values 
plt.scatter(mlnew_pred,ml_new.resid_pearson,c="r"),plt.axhline(y=0,color='blue');plt.xlabel("fitted_values");plt.ylabel("residuals")
plt.show()
```

```python
# histogram
plt.hist(ml_new.resid_pearson) # Checking the standardized residuals are normally distributed
plt.show()
```

```python
# QQ plot for residuals 
import pylab          
import scipy.stats as st

# Checking Residuals are normally distributed
st.probplot(ml_new.resid_pearson, dist="norm", plot=pylab)
plt.show()
```

```python
### Splitting the data into train and test data 

from sklearn.model_selection import train_test_split
tcar_train,tcar_test  = train_test_split(tcar_new,test_size = 0.2) # 20% size

# preparing the model on train data 

model_train = smf.ols("Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight",data=tcar_train).fit()

# train_data prediction
train_pred = model_train.predict(tcar_train)

# train residual values 
train_resid  = train_pred - tcar_train.Price

# RMSE value for train data 
train_rmse = np.sqrt(np.mean(train_resid*train_resid))
print(train_rmse)
```

```python
# prediction on test data set 
test_pred = model_train.predict(tcar_test)

# test residual values 
test_resid  = test_pred - tcar_test.Price

# RMSE value for test data 
test_rmse = np.sqrt(np.mean(test_resid*test_resid))
print(test_rmse)
```
